```python
import matplotlib.pyplot as plt

import numpy as np

import csv

from matplotlib import ticker

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
# Graph - 1

x = []
y = []

with open('gold_prices.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(row[0])
        y.append(int(row[1]))
        
plt.figure(figsize=(15, 5))
plt.plot(x,y, 'o-', label='Gold price in USD')

plt.title('Gold price from 2000 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Price in usd', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()
```

```python
# Graph - 2

x = []
y1 = []
y2 = []

with open('gold_and_platinum_prices.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(row[0])
        y1.append(int(row[1]))
        y2.append(int(row[2]))
        
plt.figure(figsize=(15, 5))
plt.plot(x,y1, 'o-', label='Gold price in USD')
plt.plot(x,y2, 'o-', label='Platinum price in USD')

plt.title('Gold and Platinum price from 2000 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Price in usd', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()
```
